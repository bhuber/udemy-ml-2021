```python
import torch
import torch.nn as nn
import torch.nn.functional as F
torch.__version__
```

```python
class Model(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.cnns = nn.ModuleList(
            [nn.Conv2d(channel_in, channel_out, kernel_size=3, stride=2)
            for channel_in, channel_out in zip(channels[:-1], channels[1:])]
        )
        self.top = nn.Linear(40*2*2, 10)

    def forward(self, x):
        # breakpoint()
        for layer in self.cnns:
            x = F.relu(layer(x))

        return self.top(x.reshape(len(x), -1))    
```

```python
model = Model([1, 10, 20, 40])
model(torch.randn(2, 1, 28, 28)).shape
```
